## np.arange(start, stop, step)

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0, 5, 0.1)
y=np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0, 5, 10)
y=np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N =8
y = np.zeros(N)

x1 = np.linspace(0 , 10, N, endpoint=True)# 10을 포함해서 8개의 수로 등분, 전체7등분
x2 = np.linspace(0 , 10, N, endpoint=False)# 10을뻬고 수로 등분, 전체8등분

plt.plot(x1, y+0.5, 'o', label='endpoint=True')
plt.plot(x2, y, 'o', label='endpoint=False')

plt.legend()
plt.ylim([-0.5, 1])# y 축 범위 지정
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N =8
y = np.zeros(N)

x1 = np.linspace(0 , 10, N, endpoint=True)# 10을 포함해서 8개의 수로 등분, 전체7등분
x2 = np.linspace(0 , 10, N, endpoint=False)# 10을뻬고 수로 등분, 전체8등분

plt.plot(x1, y+0.5, 'o-', label='endpoint=True')
plt.plot(x2, y, 'o:', label='endpoint=False')

plt.legend()
plt.ylim([-0.5, 1])# y 축 범위 지정
plt.show()
```
